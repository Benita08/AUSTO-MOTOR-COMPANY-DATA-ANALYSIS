* From the above graph we can clearly see the existence of outliers.
* Lets remove the outlier using IQR method

```python
#calculate Q1(25th percentile) and Q3(75th percentile)
Q1= df['No_of_Dependents'].quantile(0.25)
Q3= df['No_of_Dependents'].quantile(0.75)
```

```python
#calculate the interquartile range(IQR)
IQR = Q3 - Q1
```

```python
#define outlier threshold
lower_bound = Q1 - 1.5* IQR
upper_bound = Q3 + 1.5* IQR
```

```python
# Remove the outliers
df.loc[(df['No_of_Dependents'] < lower_bound) | (df['No_of_Dependents'] > upper_bound), 'No_of_Dependents']=np.nan
```

```python
print("Data after removing outliers:\n", df)
```

```python
plt.title('Boxplot of No_of_Dependents after removing outliers')
sns.boxplot(data=df,x='No_of_Dependents')
```

```python
* outliers are removed.
* Now the dependents are from 1 to 4.
* Most of the customers have 2 to 3 dependents and they are equal in count
```
